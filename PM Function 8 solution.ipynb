```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
from sklearn.gaussian_process import GaussianProcessRegressor
import pandas as pd
```

```python
X = np.load('initial_data/function_8/initial_inputs.npy')
Y = np.load('initial_data/function_8/initial_outputs.npy')
```

```python
my_dict = {
    'X1': X[:,0],
    'X2': X[:,1],
    'X3': X[:,2],
    'X4': X[:,3],
    'X5': X[:,4],
    'X6': X[:,5],
    'X7': X[:,6],
    'X8': X[:,7],
    'Y' : Y
}

df = pd.DataFrame(my_dict)
df
```

```python
gpr = GaussianProcessRegressor()
gpr.fit(X, Y)

# to optimize the acquisition function, we will use gridsearch over 390,625 grid points
x1 = np.linspace(0, 1, 5)
x2 = np.linspace(0, 1, 5)
x3 = np.linspace(0, 1, 5)
x4 = np.linspace(0, 1, 5)
x5 = np.linspace(0, 1, 5)
x6 = np.linspace(0, 1, 5)
x7 = np.linspace(0, 1, 5)
x8 = np.linspace(0, 1, 5)

X_grid = []
for i in range(len(x1)):
    for j in range(len(x2)):
        for k in range(len(x3)):
            for l in range(len(x4)):
                for m in range(len(x5)):
                    for n in range(len(x6)):
                        for o in range(len(x7)):
                            for p in range(len(x8)):
                                X_grid.append([x1[i], x2[j], x3[k], x4[l], x5[m], x6[n], x7[o], x8[p]])

X_grid = np.array(X_grid)
mean, std = gpr.predict(X_grid, return_std = True)
ucb = mean + 1.96 * std

idx_max = np.argmax(ucb)
next_query = X_grid[idx_max]
print(str(round(next_query[0],6)) + '-' + str(round(next_query[1],6)) + '-' + str(round(next_query[2],6)) + '-' + str(round(next_query[3],6)) + '-' + str(round(next_query[4],6)) + '-' + str(round(next_query[5],6)) + '-' + str(round(next_query[6],6)) + '-' + str(round(next_query[7],6)))
```
